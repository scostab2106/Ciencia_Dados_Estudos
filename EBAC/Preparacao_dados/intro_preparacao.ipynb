```python
import pandas as pd
```

Leitura do Data Frame com o Pandas e armazenamento na variável df

```python
df = pd.read_csv(r'C:\Users\samco\OneDrive\LP\Repositorios DEV\EBAC_Estudos\EBAC\Preparacao_dados\clientes-v2.csv')
```

Exibir os primeiros e os últimos registros do dataframe em string e editar o formato da data

```python
print(df.head().to_string())
print(df.tail().to_string())
df['data'] = pd.to_datetime(df['data'], format = '%d/%m/%Y', errors='coerce')
```

Mostrar as informações do dataframe

```python
print('Verificação inicial: \n')
print(df.info())
```

Verificação de dados nulos

```python
#Somando os dados nulos e exibindo a quantidade
print('Analise de dados nulos: \n', df.isnull().sum())

#Exibindo a porcentagem dos dados nulos
print("\n%"' de dados nulos: \n', df.isnull().mean() * 100)
```

Remoção de dados nulos

```python
#Removendo dados nulos
df.dropna(inplace=True)

print('Confirmar remoção de dados nulos, nulos: \n', df.isnull().sum().sum())
```

Verificação de duplicatas

```python
print('Dados duplicados: \n', df.duplicated().sum())
```

Verificação de dados únicos

```python
#Possível verificar chaves primárias
print('Dados únicos: \n', df.nunique())
```

Mostrar estatísticas dos dados

```python
print('Estatísticas: \n', df.describe())
```

Reduzir para a quantidade de colunas do meu interesse

```python
df = df[['idade','data','estado','salario','nivel_educacao','numero_filhos','estado_civil','area_atuacao']]
print(df.head().to_string())
```
