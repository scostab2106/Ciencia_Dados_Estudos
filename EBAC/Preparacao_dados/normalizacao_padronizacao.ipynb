Deletar campos que não vao ser usados

```python
#df = df.drop(labels=['data', 'estado', 'nivel_educacao', 'numero_filhos', 'area_atuacao'], axis=1) #pode fazer desse jeito ou...
df = df[['idade', 'salario']]#desse jeito
```

Normalização - MinMaxScaler

```python
scaler = MinMaxScaler()

#Normalizar todo mundo no padrão entre 0 - 1
#df['idadeMinMaxScaler'] = scaler.fit_transform(df[['idade']]) #Criando novo campo idade para comparação
#df['salarioMinMaxScaler'] = scaler.fit_transform(df[['salario']])#Criando novo campo salario para comparação

#Mudando o padão para -1 - 1
min_max_scaler = MinMaxScaler(feature_range=(-1, 1))
df['idadeMinMaxScaler'] = scaler.fit_transform(df[['idade']]) #Criando novo campo idade para comparação
df['salarioMinMaxScaler'] = scaler.fit_transform(df[['salario']])#Criando novo campo salario para comparação

print(df)
```

Padronização - StandartScaler

```python
scaler = StandardScaler()
df['idadeStandartScaler'] = scaler.fit_transform(df[['idade']])
df['salarioStandartScaler'] = scaler.fit_transform(df[['salario']])

print(df)
```

Padronização - RobustScaler

```python
scaler = RobustScaler()
df['idadeRobustScaler'] = scaler.fit_transform(df[['idade']])
df['salarioRobustScaler'] = scaler.fit_transform(df[['salario']])

print(df)
```

Exibir as métricas de maneira formatada para melhor visualização

```python
print('MinMaxScaler (de -1 a 1):\n')
print('Idade - Min: {:.4f} Max: {:.4f} Mean: {:.4f} Std: {:.4f}'.format(df['idadeMinMaxScaler'].min(), df['idadeMinMaxScaler'].max(), df['idadeMinMaxScaler'].mean(), df['idadeMinMaxScaler'].std())) #{:.4f} = Exibe até 4 casas decimais / Mean = Media, Std = Desvio Padrão
print('Salário - Min: {:.4f} Max: {:.4f} Mean: {:.4f} Std: {:.4f}\n'.format(df['salarioMinMaxScaler'].min(), df['salarioMinMaxScaler'].max(), df['salarioMinMaxScaler'].mean(), df['salarioMinMaxScaler'].std())) 
print('StandartScaler (de -1 a 1):\n')
print('Idade - Min: {:.4f} Max: {:.4f} Mean: {:.18f} Std: {:.4f}'.format(df['idadeStandartScaler'].min(), df['idadeStandartScaler'].max(), df['idadeStandartScaler'].mean(), df['idadeStandartScaler'].std())) 
print('Salário - Min: {:.4f} Max: {:.4f} Mean: {:.18f} Std: {:.4f}\n'.format(df['salarioStandartScaler'].min(), df['salarioStandartScaler'].max(), df['salarioStandartScaler'].mean(), df['salarioStandartScaler'].std())) 
print('RobustScaler (Ajusta a mediana e IQR):\n')
print('Idade - Min: {:.4f} Max: {:.4f} Mean: {:.4f} Std: {:.4f}'.format(df['idadeRobustScaler'].min(), df['idadeRobustScaler'].max(), df['idadeRobustScaler'].mean(), df['idadeRobustScaler'].std())) 
print('Salário - Min: {:.4f} Max: {:.4f} Mean: {:.4f} Std: {:.4f}'.format(df['salarioRobustScaler'].min(), df['salarioRobustScaler'].max(), df['salarioRobustScaler'].mean(), df['salarioRobustScaler'].std())) 
```
