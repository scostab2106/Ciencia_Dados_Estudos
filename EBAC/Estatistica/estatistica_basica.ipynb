```python
import pandas as pd
```

```python
df = pd.read_csv("ESPORTISTAS_MOD7.csv", delimiter=";")
df.head(25)
```

```python
df.groupby('Esporte')['Atleta'].count().reset_index().sort_values(by='Atleta', ascending=False)
# Groupby agrupa os dados do df pela coluna 'Esporte
# O count() conta o número de ocorrências da coluna 'Atleta'
# O reset_index() reseta o índice do resultado anterior
# O sort_values ordena os resultados do DataFrame pelo número de atletas
```

Trazendo os valores em porcentagem

```python
contagem_por_esporte = df['Esporte'].value_counts(normalize=True).reset_index()
# O parâmetro normalize=True faz com que os resultados sejam normalizados, ou seja, as contagens são divididas pelo número total de entradas, para fornecer a proporção de cada esporte em relação ao total.
contagem_por_esporte
```

MÉDIA 1

```python
soma_ganhos = df['Ganhos_Totais'].sum() #soma da coluna 'Ganhos_Totais'
qtde_resgistros = len(df['Ganhos_Totais']) #quantidade de atletas, ou seja, quantidade de registros na coluna 'Ganhos_Totais'
media_ganhos = soma_ganhos / qtde_resgistros #cálculo da média

print("A média de ganhos totais dos atletas é:",media_ganhos)
```

MÉDIA 2 (FUNÇÃO)

```python
df['Ganhos_Totais'].mean() #média dos ganhos totais dos atletas usando o método mean() do pandas
```

MÉDIA DOS GANHOS TOTAIS POR ESPORTE

```python
df.groupby('Esporte')['Ganhos_Totais'].mean().reset_index().sort_values(by='Ganhos_Totais', ascending=False)
```

MEDIANA TOTAL

```python
df['Ganhos_Totais'].median()
```

MEDIANA POR ESPORTE

```python
df.groupby('Esporte')['Ganhos_Totais'].median().reset_index().sort_values(by='Ganhos_Totais', ascending=False)
```

MODA

```python
df['Ganhos_Totais'].mode()
```

DESVIO PADRÃO

```python
desvio_padrao_por_esporte = df.groupby('Esporte')['Ganhos_Totais'].std().reset_index()

desvio_padrao_por_esporte

# Não tem como calcular desvio padrão de valor que aparece só uma vez
    
```
