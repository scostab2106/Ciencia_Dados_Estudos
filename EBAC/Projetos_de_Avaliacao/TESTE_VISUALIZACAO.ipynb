```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns   
```

```python
df = pd.read_csv('ecommerce_preparados.csv')
df.head()
```

Histograma

```python
plt.figure(figsize=(10, 6))
df['Gênero'].value_counts().plot(kind='bar', color='#90ee70')
plt.xlabel('Gênero')
plt.ylabel('Quantidade')
plt.xticks(rotation=90) 
sns.despine() #Remover as bordas do gráfico
plt.show()
```

Dispersão

```python
plt.scatter(df['Preço'], df['N_Avaliações'], alpha=0.5, color='#278f65')
plt.colorbar(label = 'Contagem dentro do bin')
plt.xlabel('Preço')
plt.ylabel('Avaliação')
plt.title('Dispersão de Preço e Avaliação')
plt.show()
sns.despine()
```

Calor

```python
df_corr = df[['Nota','N_Avaliações','Desconto','Marca_Cod','Material_Cod','Temporada_Cod','Qtd_Vendidos_Cod','Preço']].corr() 
plt.figure(figsize=(10, 8))
sns.heatmap(df_corr, annot=True, fmt=".2f")
plt.title('Mapa de Calor da Correlação entre os Atributos')
plt.show()
```

Barra

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Nota'], df['N_Avaliações'])
plt.title('Dispersão - Número de avaliações x Nota')
plt.ylabel('Número de Avaliações')
plt.xlabel('Nota')
sns.despine()
```

Pizza

```python

x = df['Gênero'].value_counts().index 
y = df['Gênero'].value_counts().values 
plt.figure(figsize=(30,30))
plt.pie(y, labels=x, autopct='%.1f%%', startangle = 90)
plt.title('Distribuição de Gêneros')
sns.despine()
plt.show()
```

Densidade

```python
plt.figure(figsize=(10,6))
sns.kdeplot(df['Nota'], fill=True, color='#863e9c')
plt.title('Densidade de Notas')
plt.xlabel('Nota')
sns.despine()
plt.show()
```

Regressão

```python
sns.regplot(x='Qtd_Vendidos_Cod', y='Preço', data=df)
sns.despine()
plt.title('Regressão - Quantidade Vendida x Preço')
plt.ylabel('Preço')
plt.xlabel('Quantidade Vendida')
plt.show()
```
