```python
import matplotlib.pyplot as plt
import pandas as pd     
```

```python
df = pd.read_csv('clientes-v3-preparado.csv')
print(df.head(20).to_string())  
```

Gráfico de barras

```python
plt.figure(figsize=(10, 6))
df['nivel_educacao'].value_counts().plot(kind='bar', color='#90ee70')#value_counts() = função para obter informações do campo definido do dataframe, .plot = função do pandas para visualizar os valores passando como parâmetros o tipo do gráfico e a cor  
plt.title('Divisão de Escolaridade - 1')
plt.xlabel('Nível de Educação')
plt.ylabel('Quantidade')
plt.xticks(rotation=0) #Fixa a legenda do eixo x em 0 graus
plt.show()


x = df['nivel_educacao'].value_counts().index #Define quem é o eixo x e usa o value counts pra pegar o índice
y = df['nivel_educacao'].value_counts().values #Define o eixo y e usa o value counts pra pegar os valores
plt.figure(figsize=(10, 6))
plt.bar(x, y, color = '#60aa65')
plt.title('Divisão de Escolaridade - 2')
plt.xlabel('Nível de Educação')
plt.ylabel('Quantidade')
```

Gráfico de Pizza

```python
plt.figure(figsize=(10, 6))
plt.pie(y, labels=x, autopct='%.2f%%', startangle = 90)#indicar qual vai ser o label / autopct informa quantas casas decimais vão ser mostradas / startangle diz qual vai ser o ângulo de inicio das divisoes.
plt.title('Distribuição de Nível de Educação')
plt.show()
```

Gráfico de dispersão

```python
plt.hexbin(df['idade'], df['salario'], gridsize=40, cmap='Blues') #cmap é um mapa de cores, nesse caso o mapa de azul
plt.colorbar(label = 'Contagem dentro do bin')
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.title('Dispersão de Idade e Salário')
plt.show()
```
