```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('clientes-v3-preparado.csv')
print(df.head())
```

Heatmap de correlação

```python
df_corr = df[['salario','idade','anos_experiencia','numero_filhos', 'nivel_educacao_cod', 'area_atuacao_cod', 'estado_cod']].corr() # Informando os campos a serem feitas as correlações
plt.figure(figsize=(10, 8))
sns.heatmap(df_corr, annot=True, fmt=".2f")
plt.title('Mapa de Calor da Correlação entre Variáveis')
sns.despine() #Remover as bordas do gráfico
plt.show()
```

Countplot

```python
sns.countplot(x='estado_civil', data=df)
plt.title('Distribuição do Estado Civil')
plt.xlabel('Estado Civil')
plt.ylabel('Contagem')
sns.despine() #Remover as bordas do gráfico
plt.show()
```

Countplot com legenda

```python
sns.countplot(x='estado_civil', hue='nivel_educacao', data=df)
plt.title('Distribuição do Estado Civil')
plt.xlabel('Estado Civil')
plt.ylabel('Contagem')
sns.despine() #Remover as bordas do gráfico
plt.show()
```
