```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\samco\OneDrive\LP\Repositorios DEV\EBAC_Estudos\EBAC\Visualizacao_dados\clientes-v3-preparado.csv')
print(df.head().to_string())
```

HISTOGRAMAS

```python
plt.hist(df['salario']) #Gerando histograma da coluna salario
plt.show()#Exibindo o gráfico
```

Criando parâmetros para o histograma

```python
plt.figure(figsize=(10, 6)) #Tamanho da figura
plt.hist(df['salario'], bins=100, color='green', alpha=0.8) #bins é o número de barras, color é a cor das barras e alpha é a transparência
plt.title('Histograma - Distribuição de salários') #Título do gráfico
plt.xlabel('Salário') #Rótulo do eixo x
plt.xticks(ticks=range(0, int(df['salario'].max())+2000, 2000)) #Definindo os ticks do eixo x para melhor visualização, indo de 0 até o valor máximo de salário arredondado para cima, com intervalos de 2000
plt.ylabel('Frequência') #Rótulo do eixo y
plt.grid(True) #Exibe as grades do gráfico
plt.show()
```

Múltiplos gráficos

```python
plt.figure(figsize=(10, 6)) #Tamanho da figura
plt.subplot(2, 2, 1) #2 Linhas, 2 Colunas, 1º Gráfico

#Gráfico de dispersão
plt.scatter(df['salario'], df['salario'])
plt.title('Dispersão - Salário e Salário')
plt.xlabel('Salário')
plt.ylabel('Salário')

plt.subplot(1, 2, 2) #1 Linha, 2 Colunas, 2º Gráfico
plt.scatter(df['salario'], df['anos_experiencia'], color='#5883a8', alpha=0.6, s=30) #s=30 é o tamanho da bola
plt.title('Dispersão - Salário e Anos de Experiência')
plt.xlabel('Salário')
plt.ylabel('Anos de Experiência')

#Mapa de calor com o seaborn dentro do matplotlib
corr = df[['salario', 'anos_experiencia']].corr() #Calculando a correlação entre as colunas salario e anos_experiencia
plt.subplot(2, 2, 3) #2 Linhas, 2 Colunas, 3º Gráfico
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlação Salário e Anos de Experiência')

plt.tight_layout() #Ajusta o layout para evitar sobreposição dos gráficos e espaçamentos
plt.show()
```
